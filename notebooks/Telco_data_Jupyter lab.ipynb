```python
import pandas as pd
```

```python
df=pd.read_csv("Telco-Customer-Churn.csv")
```

```python
df.info()
```

```python
#check missing value
df.isnull().sum() #is is necessary because ML model can not work with blank string
```

```python
#check blank values
(df == " ").sum()
```

```python
#replace blank values with NaN
import numpy as np
df["TotalCharges"]=df["TotalCharges"].replace("",np.nan)
```

```python
#convert totalcharges (string) into numeric
df["TotalCharges"]=pd.to_numeric(df["TotalCharges"],errors="coerce")
```

```python
df["TotalCharges"].fillna(df["TotalCharges"].median())
```

```python
#check duplicate records
df.duplicated().sum()
```

```python
#remove customer id column because its not necessary to predict churn
df.drop("customerID",axis=1,inplace=True)
```

```python
#check datatypes
df.dtypes  #ML algo require numerical col to have numeric data type
```

### encode categorical variable
#### most column contain text values such as male/female,Yes/no,Month-to-month,Dsl,convert them numeric values

#### ML model generally require numeric input 

```python
from sklearn.preprocessing import LabelEncoder#(class)
le = LabelEncoder()#(object)
for column in df.select_dtypes(include="object").columns:
    df[column]=le.fit_transform(df[column].astype(str))
```
